```
 zip_to_latlong = {}
 for i in sat_to_zip.iterrows():
   zip_to_latlong[i[1]['zip']] = (i[1]['latitude'], i[1]['longitude'])
 zip_to_latlong
```

```
tile_bounds = {}
for tile in tiles:
  latitude, longitude = latlon(tile[0], tile[1], 14)
  latitude2, longitude2 = latlon(tile[0] + 1, tile[1] + 1, 14)
  tile_bounds[tile] = (latitude, longitude, latitude2, longitude2)
tile_bounds
```

```
def zipcode_in_tile(zipcode):
  latitude, longitude = zip_to_latlong[zipcode]
  for tile in tiles:
    if latitude > tile_bounds[tile][2] and latitude < tile_bounds[tile][0] and longitude > tile_bounds[tile][1] and longitude < tile_bounds[tile][3]:
      return tile
  return None
```

```
zip_to_tile = {}
remaining_tile_set = set(tiles)
for row in census_data.iterrows():
  zip = int(row[1]['ZIPCODE'])
  if zip not in zip_to_latlong:
    continue
  tile = zipcode_in_tile(zip)
  if tile is not None:
    zip_to_tile[zip] = [tile]
    if tile in remaining_tile_set:
      remaining_tile_set.remove(tile)
zip_to_tile
```

```
from util import getElevation

for tile in remaining_tile_set:
  if getElevation(tile_bounds[tile][0], tile_bounds[tile][1]) != 0:
    tile_center_lat = (tile_bounds[tile][0] + tile_bounds[tile][2]) / 2
    tile_center_long = (tile_bounds[tile][1] + tile_bounds[tile][3]) / 2
    closest_distance = float("inf")
    closest_zip = None
    for zip in zip_to_latlong:
      dist = (tile_center_lat - zip_to_latlong[zip][0]) ** 2 + (tile_center_long - zip_to_latlong[zip][1]) ** 2
      if dist < closest_distance:
        closest_distance = dist
        closest_zip = zip
    if closest_zip in zip_to_tile:
      zip_to_tile[closest_zip].append(tile)
    else:
      zip_to_tile[closest_zip] = [tile]
zip_to_tile
```
